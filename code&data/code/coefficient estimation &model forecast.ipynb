```python
# 滚动生成预测模型

# 定义计算回归参数的函数
lag = 1

def ols_params(df, xcols,  ycol):
    return sm.OLS(df[ycol], sm.add_constant(df[xcols])).fit(cov_type='HAC',cov_kwds={'maxlags':lag}).params


# 定义计算Rsquared的函数
def ols_rsquared(df, xcols,  ycol):
    return sm.OLS(df[ycol], sm.add_constant(df[xcols])).fit(cov_type='HAC',cov_kwds={'maxlags':lag}).rsquared

# 定义回归参数显著性的函数(p)
def ols_pvalue(df, xcols,  ycol):
    return sm.OLS(df[ycol], sm.add_constant(df[xcols])).fit(cov_type='HAC',cov_kwds={'maxlags':lag}).pvalues

# 定义回归参数显著性的函数(t)
def ols_tvalue(df, xcols,  ycol):
    return sm.OLS(df[ycol], sm.add_constant(df[xcols])).fit(cov_type='HAC',cov_kwds={'maxlags':lag}).tvalues
```
